```python
from qiskit import QuantumCircuit 
from qiskit_aer import AerSimulator
from qiskit.compiler import transpile
# Runtime for Basis is O(MN) N = number of qubits
# Dense angle encoding, can encode 2 feature 2 qubits 

#Blue - XX 
#Green - XY
#Red - YY

qc = QuantumCircuit(3)
theta = (3.14 / 4)

#Encode
qc.ch(0,1)
qc.ch(1,2)
qc.ry(theta,0, label="Blue_Qubit")
qc.ry(theta,1, label="Red_Qubit")
qc.ry(theta,2, label="Green_Qubit")

# Variational Model
qc.cx(0,1)
qc.cx(0,2)
qc.cx(1,2)

simulator = AerSimulator()

print("Available simulation devices:", simulator.available_devices() )

compiled_cir = transpile(qc, simulator)
result = simulator.run(compiled_cir).result()
```
