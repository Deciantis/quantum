```python
from qiskit import QuantumCircuit
# Runtime for Basis is O(MN) N = number of qubits
# Dense angle encoding, can encode 2 feature 2 qubits 

#Blue - XX 
#Green - XY
#Red - YY



qc = QuantumCircuit(3)
theta = (3.14 / 4)

#Encode
qc.ch(0,1)
qc.ch(1,2)

qc.ry(theta,0, label="Blue_Qubit")
qc.ry(theta,1, label="Red_Qubit")
qc.ry(theta,2, label="Green_Qubit")
```

```python
qc.draw("mpl")
```
